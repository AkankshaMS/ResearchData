Calculate and add Corrected height in cm as a column to dataframe

```
temp = np.modf(df.Height)
feetHeight = temp[1].values*30.48
inchHeight = temp[0]*2.54
totalHeight = tuple(map(lambda i, j: i + j, feetHeight, inchHeight))

df['CorrHeight'] = totalHeight
```

Calculate and add BMI as a separate column to the dataframe using weight in kg and height in cm

```
df['BMI'] = ((df.Weight)/(df.CorrHeight*df.CorrHeight))*10000
df.head(5)
```

```
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV, StratifiedKFold
SEED = 42
import warnings
warnings.filterwarnings("ignore")

## Baseline model performance evaluation
# kf = KFold(n_splits=5, shuffle=True, random_state=SEED)   # this may result in imbalance classes in each fold
kf = StratifiedKFold(n_splits=4, shuffle=True, random_state=SEED)

# to give model baseline report in dataframe 
def baseline_report(model, xxx, X_test, y_train, y_test, name):
    model.fit(xxx, y_train.ravel())
    accuracy     = np.mean(cross_val_score(model, xxx, y_train, cv=kf, scoring='accuracy'))
    precision    = np.mean(cross_val_score(model, xxx, y_train, cv=kf, scoring='precision'))
    recall       = np.mean(cross_val_score(model, xxx, y_train, cv=kf, scoring='recall'))
    f1score      = np.mean(cross_val_score(model, xxx, y_train, cv=kf, scoring='f1'))
   
    y_pred = model.predict(X_test)
    

    df_model = pd.DataFrame({'model'        : [name],
                             'accuracy'     : [accuracy],
                             'precision'    : [precision],
                             'recall'       : [recall],
                             'f1score'      : [f1score],                           
                             'timetaken'    : [0]       })   # timetaken: to be used for comparison later
    return df_model

# to evaluate baseline models
gnb = GaussianNB()

logit = LogisticRegression()

randomforest = RandomForestClassifier()


# to concat all models
df_models = pd.concat([baseline_report(gnb, xxx, X_test, y_train, y_test, 'GaussianNB'),
                       baseline_report(logit,xxx, X_test, y_train, y_test, 'LogisticRegression'),
                       baseline_report(randomforest, xxx, X_test, y_train, y_test, 'RandomForest')
                       ], axis=0).reset_index()
df_models = df_models.drop('index', axis=1)
df_models
```
